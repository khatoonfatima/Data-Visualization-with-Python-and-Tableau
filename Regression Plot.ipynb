```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import math
import scipy as stats
import seaborn as sns
from numpy.random import randn
import matplotlib as mlp
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips') #Default dataset in jupyter
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips) #lmplot is the 2D scatterplot 
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,order=4,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
#For avoiding the regression plot

sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip_pect',tips)
```

```python
sns.lmplot('tip_pect','tip',tips,
          scatter_kws={'marker':'o','color':'blue'},
          line_kws={'linewidth':1,'color':'red'})
```

```python
sns.lmplot('size','tip_pect',tips,fit_reg=True) #With regression
```

```python
sns.lmplot('size','tip_pect',tips,fit_reg=False) #By Avoiding the regression
```

```python
#For adding jitter

sns.lmplot('size','tip_pect',tips,x_jitter=.1)
```

```python
#For calculating the estimated mean

sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
#Seaborn can also provide local regression
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True)
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={'color':'red'})
```

```python
sns.regplot('total_bill','tip_pect',tips) #regression plot 
```

```python
sns.regplot('total_bill','size',tips)
```

```python
fig, (axis1,axis2)=plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','tip_pect',tips,ax=axis1)

sns.violinplot(tips['tip_pect'],tips['day'],color='blue',ax=axis2)
```
